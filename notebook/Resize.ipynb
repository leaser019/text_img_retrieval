```python
import os
import shutil
from PIL import Image

# Đường dẫn thư mục chứa các folder keyframes
base_folders = [
    '/content/drive/MyDrive/video_keyframes/',
]

# Đường dẫn thư mục lưu trữ ảnh đã xử lý
output_folder = '/content/drive/MyDrive/Pipeline_HCM_AI/images'

# Tạo thư mục lưu trữ nếu chưa tồn tại
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Duyệt qua từng thư mục keyframes
for base_folder in base_folders:
    for i in range(1, 33):  # Tạo 30 folder con
        folder_name = f"L3_V{i:03}"
        images_folder = os.path.join(base_folder, folder_name)

        # Kiểm tra nếu thư mục tồn tại
        if os.path.exists(images_folder):
            for image_name in os.listdir(images_folder):
                image_path = os.path.join(images_folder, image_name)

                # Kiểm tra nếu là file ảnh
                if os.path.isfile(image_path):
                    try:
                        # Mở và chỉnh kích thước ảnh
                        with Image.open(image_path) as img:
                            resized_img = img.resize((224, 112))
                            output_path = os.path.join(output_folder, f"{folder_name}_{image_name}")
                            resized_img.save(output_path)
                    except Exception as e:
                        print(f"Không thể xử lý {image_path}: {e}")
        else:
            print(f"Thư mục không tồn tại: {images_folder}")

print("Hoàn thành xử lý ảnh.")
```
