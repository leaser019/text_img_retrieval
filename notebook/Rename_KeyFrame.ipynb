```python
import os
import pandas as pd

csv_path = '/content/drive/MyDrive/Video AI /map-keyframes/L12_V001.csv'

images_folder = '/content/drive/MyDrive/Video AI /keyframes - Video 12/L12_V001'

df = pd.read_csv(csv_path)

for index, row in df.iterrows():
    n_value = int(row['n'])
    frame_idx = int(row['frame_idx'])

    old_image_name = f"{n_value:03}.jpg"

    new_image_name = f"L12_V001_{frame_idx}.jpg"

    old_image_path = os.path.join(images_folder, old_image_name)
    new_image_path = os.path.join(images_folder, new_image_name)

    if os.path.exists(old_image_path):
        os.rename(old_image_path, new_image_path)
        print(f"Đã đổi tên: {old_image_name} -> {new_image_name}")
    else:
        print(f"Không tìm thấy: {old_image_path}")
```

```python
import os
import pandas as pd

base_images_folder = '/content/drive/MyDrive/Video AI /keyframes - Video 12'
base_csv_folder = '/content/drive/MyDrive/Video AI /map-keyframes'

for i in range(1, 31):  # 1 đến 30
    folder_name = f"L12_V{i:03}"
    csv_name = f"{folder_name}.csv"

    images_folder = os.path.join(base_images_folder, folder_name)
    csv_path = os.path.join(base_csv_folder, csv_name)

    if not os.path.exists(csv_path):
        print(f"Không tìm thấy file CSV: {csv_path}")
        continue

    df = pd.read_csv(csv_path)

    for index, row in df.iterrows():
        n_value = int(row['n'])
        frame_idx = int(row['frame_idx'])

        old_image_name = f"{n_value:03}.jpg"

        new_image_name = f"{folder_name}_{frame_idx}.jpg"

        old_image_path = os.path.join(images_folder, old_image_name)
        new_image_path = os.path.join(images_folder, new_image_name)

        if os.path.exists(old_image_path):
            os.rename(old_image_path, new_image_path)
            print(f"Đã đổi tên: {old_image_name} -> {new_image_name}")
        else:
            print(f"Không tìm thấy file ảnh: {old_image_path}")
```
